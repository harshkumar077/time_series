```python
def test_stationary(timeseries):
    rolmean=timeseries.rolling(12).mean()
    rolstd=timeseries.rolling(12).std()
    plt.figure(figsize=(18,10))
    plt.plot(timeseries)
    plt.plot(rolmean,label='rol_mean')
    plt.plot(rolstd,label='rol_std')
    plt.legend()
    plt.show()
    print("---------------------")
    print("result of argumented dicky fuller")
    dftest=adfuller(timeseries,autolag='AIC')
    dfoutput=pd.Series(dftest[0:4],index=['Test Statistics','p-value','lag_used','no_of_observ'])
    for key ,value in dftest[4].items():
        dfoutput['confidence interval(%s)'%key]=value
    print(dfoutput)
      
```

```python
test_stationary(ts_sales)
```

```python
# from sklearn.metrics import mean_squared_error,mean_absolute_error

# def arima_model(x,y,arima_order):
#     history=[ i for i in x]
#     prediction=list()
#     for i in range(len(y)):
#         model=ARIMA(history,order=arima_order)
#         model_fit=model.fit()
#         yhat=model.forecast()
#         prediction.append(yhat)
#         history.append(y[i])
#     mse=np.sqrt(mean_squared_error(y,prediction))  
#     return mse
    
import numpy as np
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

def arima_model(x, y, arima_order):
    history = [i for i in x]
    predictions = list()
    for i in range(len(y)):
        model = ARIMA(history, order=arima_order)
        model_fit = model.fit()
        
        # Use this line for forecasting with new ARIMA implementation
        yhat = model_fit.forecast(steps=1)[0]
        
        predictions.append(yhat)
        history.append(y[i])
        
    mse = np.sqrt(mean_squared_error(y, predictions))
    return mse
```

```python
def evaluate_arima_model(data,test,p_value,d_value,q_value):
    data=data.astype('float32')
    best_score,best_cfg=float('inf'),None
    for i in p_value:
            for j in d_value:
                for k in q_value:
                    order=(i,j,k)
                    try:
                        rmse=arima_model(data,test,order)
                        if rmse<best_score:
                            best_score,best_cfg=rmse,order
                        print("ARIMAA%s, RMSE=%.3f" %(order,rmse))
                    except Exception as e:
                            print(e)
                            print("errroe")
                            continue;
    print(f"best_rmse and best_confg {best_score}{best_cfg}")
```
